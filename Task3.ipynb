## 2. Data Preprocessing

Steps performed:
- Converted date column (dteday) to datetime and extracted features (day, month, weekday, year, etc.)
- Encoded categorical variables (season, weathersit, etc.) using one-hot encoding
- Handled numerical features by scaling/normalization if needed
- Checked for and handled missing values

This ensures our dataset is ready for machine learning models.


```python

df["dteday"] = pd.to_datetime(df["dteday"])
df["yr"] = df["dteday"].dt.year
df["mnth"] = df["dteday"].dt.month
df["day"] = df["dteday"].dt.day

df.head()
```

```python

df["is_weekend"] = df["weekday"].apply(lambda x: 1 if x >= 5 else 0)

categorical_cols = ["season", "weathersit", "mnth", "weekday"]

df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

df.head()
```

## 4. Model Building

We test multiple regression models to predict cnt:
- Linear Regression
- Decision Tree Regressor
- Random Forest Regressor
- Gradient Boosting Regressor

  
The motivation is to compare simple vs. complex models and evaluate trade-offs between interpretability and performance.


```python
from sklearn.model_selection import train_test_split

X = df.drop(["cnt", "dteday","registered","casual","instant"], axis=1)
y = df["cnt"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=7
)

df.head()
```

## 5. Model Evaluation

We use the following metrics:
- Mean Absolute Error (MAE)
- Mean Squared Error (MSE)
- Root Mean Squared Error (RMSE)
- R² Score

Cross-validation is used to ensure robust performance estimates.  
We compare models on the basis of both accuracy and generalization

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

model = LinearRegression()

model.fit(X_train,y_train)

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R² Score:", r2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

coef_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Coefficient': model.coef_
})

coef_df = coef_df.reindex(coef_df['Coefficient'].abs().sort_values(ascending=False).index)


colors = ['green' if c > 0 else 'red' for c in coef_df['Coefficient']]


plt.figure(figsize=(9,6))
bars = plt.barh(coef_df['Feature'], coef_df['Coefficient'], color=colors)


for bar, coef in zip(bars, coef_df['Coefficient']):
    plt.text(
        bar.get_width() + (0.02 if coef > 0 else -0.02),
        bar.get_y() + bar.get_height()/2,
        f"{coef:.3f}",
        va='center',
        ha='left' if coef > 0 else 'right',
        fontsize=10
    )

plt.axvline(0, color='black', linewidth=1)
plt.xlabel('Coefficient Value', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.title('Feature Influence on Bike Demand', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score


ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)

y_pred_ridge = ridge.predict(X_test)

mse_ridge = mean_squared_error(y_test, y_pred_ridge)
rmse_ridge = np.sqrt(mse_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print(f"Ridge Regression RMSE: {rmse_ridge:.3f}")
print(f"Ridge Regression R2 Score: {r2_ridge:.3f}")

y_pred_lr = model.predict(X_test)
print(f"Linear Regression RMSE: {np.sqrt(mean_squared_error(y_test, y_pred_lr)):.3f}")
print(f"Linear Regression R2 Score: {r2_score(y_test, y_pred_lr):.3f}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score


rf = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)
rf.fit(X_train, y_train)


y_pred_rf = rf.predict(X_test)


rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
r2_rf = r2_score(y_test, y_pred_rf)

print(f"Random Forest RMSE: {rmse_rf:.3f}")
print(f"Random Forest R2 Score: {r2_rf:.3f}")

y_pred_lr = model.predict(X_test)
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
r2_lr = r2_score(y_test, y_pred_lr)

print(f"Linear Regression RMSE: {rmse_lr:.3f}")
print(f"Linear Regression R2 Score: {r2_lr:.3f}")
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, random_state=42)
}

cv_results = {}


for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
    cv_results[name] = {
        "Mean R2": np.mean(scores),
        "Std R2": np.std(scores)
    }

import pandas as pd
cv_df = pd.DataFrame(cv_results).T
print(cv_df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor
import numpy as np

X = df.drop(columns=["cnt", "casual", "registered", "dteday"])
y_casual = df["casual"]
y_registered = df["registered"]


X_train, X_test, y_casual_train, y_casual_test, y_registered_train, y_registered_test = train_test_split(
    X, y_casual, y_registered, test_size=0.2, random_state=42
)

casual_model = GradientBoostingRegressor(random_state=42)
casual_model.fit(X_train, y_casual_train)
casual_pred = casual_model.predict(X_test)

registered_model = GradientBoostingRegressor(random_state=42)
registered_model.fit(X_train, y_registered_train)
registered_pred = registered_model.predict(X_test)

cnt_pred = casual_pred + registered_pred
y_cnt_test = y_casual_test + y_registered_test

mae = mean_absolute_error(y_cnt_test, cnt_pred)
mse = mean_squared_error(y_cnt_test, cnt_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_cnt_test, cnt_pred)

print("Brownie Points Model (Casual + Registered → cnt):")
print(f"MAE  : {mae:.2f}")
print(f"MSE  : {mse:.2f}")
print(f"RMSE : {rmse:.2f}")
print(f"R²   : {r2:.2f}")
```
