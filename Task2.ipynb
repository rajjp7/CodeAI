```python
import matplotlib.pyplot as plt

# Handle missing values
df['extras'] = df['extras'].fillna(0)
df['batsman_runs'] = df['batsman_runs'].fillna(0)

# Runs conceded = batsman runs + extras
df['runs_conceded'] = df['batsman_runs'] + df['extras']

# Wickets credited to bowler (exclude run outs, retired hurt, etc.)
wicket_kinds = ["bowled", "caught", "lbw", "stumped", "caught and bowled", "hit wicket"]
df['is_wicket'] = df['dismissal_kind'].isin(wicket_kinds).astype(int)

# Aggregate bowler stats
bowler_stats = df.groupby('bowler').agg(
    balls_bowled=('ball', 'count'),
    runs_conceded=('runs_conceded', 'sum'),
    wickets=('is_wicket', 'sum')
).reset_index()

# Derived metrics
bowler_stats['overs'] = bowler_stats['balls_bowled'] / 6
bowler_stats['economy'] = bowler_stats['runs_conceded'] / bowler_stats['overs']
bowler_stats['average'] = bowler_stats['runs_conceded'] / bowler_stats['wickets'].where(bowler_stats['wickets'] > 0)
bowler_stats['strike_rate'] = bowler_stats['balls_bowled'] / bowler_stats['wickets'].where(bowler_stats['wickets'] > 0)

# Filter bowlers with at least 200 balls bowled
filtered_bowlers = bowler_stats[bowler_stats['balls_bowled'] >= 200]

# Ranking formula (you can tweak weights as you like)
filtered_bowlers['rank_score'] = (
    filtered_bowlers['wickets']*0.5 -
    filtered_bowlers['economy']*0.3 -
    filtered_bowlers['strike_rate']*0.2
)

# Top 10 bowlers
top_10 = filtered_bowlers.sort_values(by='rank_score', ascending=False).head(10)

print("Top 10 Bowlers:")
print(top_10[['bowler', 'balls_bowled', 'wickets', 'economy', 'average', 'strike_rate']])

# --- Visualization: Wickets vs Economy ---
plt.figure(figsize=(10,6))
plt.scatter(top_10['wickets'], top_10['economy'], color='teal', s=100)

# Annotate bowler names
for i, row in top_10.iterrows():
    plt.text(row['wickets']+0.3, row['economy'], row['bowler'], fontsize=9)

plt.xlabel("Wickets")
plt.ylabel("Economy Rate")
plt.title("Top 10 IPL Bowlers (Wickets vs Economy)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

```

```python
# Pick batsman name (change this to whoever you want to analyze)
batsman_name = "V Kohli"

# Filter data: only deliveries faced by this batsman against top 10 bowlers
batsman_vs_bowlers = df[(df['batsman'] == batsman_name) & (df['bowler'].isin(top_10['bowler']))]

# Group by bowler
batsman_stats = batsman_vs_bowlers.groupby('bowler').agg(
    runs=('batsman_runs', 'sum'),
    balls=('ball', 'count'),
    dismissals=('player_dismissed', lambda x: (x == batsman_name).sum())
).reset_index()

# Calculate strike rate and average
batsman_stats['strike_rate'] = (batsman_stats['runs'] / batsman_stats['balls']) * 100
batsman_stats['average'] = batsman_stats['runs'] / batsman_stats['dismissals'].replace(0, pd.NA)

print(f"Stats of {batsman_name} against Top 10 Bowlers:")
print(batsman_stats[['bowler', 'runs', 'balls', 'strike_rate', 'dismissals', 'average']])

# --- Visualization ---
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(batsman_stats['bowler'], batsman_stats['strike_rate'], color='orange')
plt.xticks(rotation=45, ha="right")
plt.ylabel("Strike Rate")
plt.title(f"{batsman_name} Strike Rate vs Top 10 Bowlers")
plt.show()
```

```python
# Choose two batsmen (make sure to use exact names from df['batsman'].unique())
batsman_1 = "V Kohli"
batsman_2 = "MS Dhoni"

# Filter dataset: only balls faced by these batsmen vs top 10 bowlers
batsmen_vs_bowlers = df[(df['batsman'].isin([batsman_1, batsman_2])) &
                        (df['bowler'].isin(top_10['bowler']))]

# Group stats by bowler & batsman
comparison_stats = batsmen_vs_bowlers.groupby(['bowler', 'batsman']).agg(
    runs=('batsman_runs', 'sum'),
    balls=('ball', 'count'),
    dismissals=('player_dismissed', lambda x: (x.isin([batsman_1, batsman_2])).sum())
).reset_index()

# Strike rate & average
comparison_stats['strike_rate'] = (comparison_stats['runs'] / comparison_stats['balls']) * 100
comparison_stats['average'] = comparison_stats['runs'] / comparison_stats['dismissals'].replace(0, pd.NA)

print(f"Comparison of {batsman_1} and {batsman_2} vs Top 10 Bowlers:")
print(comparison_stats)

# --- Visualization: Side by side bar chart ---
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(data=comparison_stats, x="bowler", y="strike_rate", hue="batsman", palette="Set2")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Strike Rate")
plt.title(f"Strike Rate Comparison: {batsman_1} vs {batsman_2} against Top 10 Bowlers")
plt.legend(title="Batsman")
plt.show()
```

```python
import pandas as pd

# Summarize stats for each batsman
summary = comparison_stats.groupby('batsman').agg(
    total_runs=('runs','sum'),
    total_balls=('balls','sum'),
    total_dismissals=('dismissals','sum')
).reset_index()

summary['strike_rate'] = (summary['total_runs'] / summary['total_balls']) * 100
summary['average'] = summary['total_runs'] / summary['total_dismissals'].replace(0, pd.NA)

# Normalize both metrics (so strike_rate & average are comparable)
summary['sr_norm'] = (summary['strike_rate'] - summary['strike_rate'].min()) / (summary['strike_rate'].max() - summary['strike_rate'].min())
summary['avg_norm'] = (summary['average'] - summary['average'].min()) / (summary['average'].max() - summary['average'].min())

# Weighted score (50% SR + 50% Avg)
summary['final_score'] = 0.5 * summary['sr_norm'] + 0.5 * summary['avg_norm']

# Print detailed results
print("📊 Against the Top 10 Bowlers:\n")
for _, row in summary.iterrows():
    print(f"- {row['batsman']}: {row['total_runs']} runs in {row['total_balls']} balls, "
          f"SR {row['strike_rate']:.2f}, Avg {row['average']:.2f} "
          f"(Dismissals: {row['total_dismissals']})")

# Decide winner
best = summary.loc[summary['final_score'].idxmax(), 'batsman']
print(f"\n✅ Conclusion: Considering both strike rate and average, **{best}** performs better against the top 10 bowlers.")
```

# **Batsmen Analysis**

1.Top batsmen on the basis of **Runs**

```python


df['dismissed_flag'] = df['player_dismissed'].notna().astype(int)

#cleared the column dismissed flag

batsman_stats = df.groupby('batsman').agg(
    runs=('batsman_runs','sum'),
    balls=('ball','count'),
    dismissals=('player_dismissed','count')
).reset_index()

batsman_stats['strike_rate'] = (batsman_stats['runs']/ batsman_stats['balls'])*100
batsman_stats['average'] = batsman_stats['runs']/ batsman_stats['dismissals'].replace(0,1)

top_batsman = batsman_stats.sort_values(by='runs',ascending=False).head(10)

top_batsman
```

**2.Top batsmen with best boundary percent**

```python

boundary_stats = df.groupby('batsman').agg(
    balls=('ball','count'),
    fours=('batsman_runs', lambda x: (x==4).sum()),
    sixes=('batsman_runs', lambda x: (x==6).sum())
).reset_index()

boundary_stats['boundary%'] = ((boundary_stats['fours'] + boundary_stats['sixes']) / boundary_stats['balls']) * 100

dot_stats = df.groupby('batsman').agg(
    balls=('ball','count'),
    dots=('batsman_runs', lambda x: (x==0).sum())
).reset_index()

dot_stats['dot%'] = (dot_stats['dots'] / dot_stats['balls']) * 100

batsman_stats = boundary_stats.merge(dot_stats[['batsman','dot%']], on='batsman', how='left')

filtered_batsman = batsman_stats[batsman_stats['balls']>=200]

filtered_batsman = filtered_batsman.sort_values(by='boundary%', ascending=False).head(10)


filtered_batsman[['batsman','balls','boundary%','dot%']]
```

3. Top batsmen with best **Powerplay**

```python

def get_phase(over):
    if over <= 6:
        return 'Powerplay'
    elif over <= 15:
        return 'Middle overs'
    else:
        return 'Death Over'

df['phase'] = df['over'].apply(get_phase)

phase_stats = df.groupby(['batsman','phase']).agg(
    phase_runs=('batsman_runs','sum'),
    phase_balls=('ball','count')
).reset_index()

phase_stats['phase_SR'] = (phase_stats['phase_runs']/phase_stats['phase_balls'])*100


best_powerplay = phase_stats[(phase_stats['phase']=="Powerplay") & (phase_stats['phase_balls']>=100) & (phase_stats['phase_runs']>=500)]


best_powerplay = best_powerplay.sort_values(by='phase_SR', ascending=False).head(10)


best_powerplay
```

4. Top batsmen in Middle overs

```python
best_middle_over = phase_stats[(phase_stats['phase']=="Middle overs") & (phase_stats['phase_balls']>=100) & (phase_stats['phase_runs']>=1000)]

best_middle_over = best_middle_over.sort_values(by='phase_SR', ascending=False).head(10)

best_middle_over
```

```python
best_death = phase_stats[(phase_stats['phase']=="Death Over") & (phase_stats['phase_balls']>=100) & (phase_stats['phase_runs']>=1000)]

best_death = best_death.sort_values(by='phase_SR', ascending=False).head(20)

best_death
```
